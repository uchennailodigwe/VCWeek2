```python
#####Feature Engineering

#1. Create a new column for Date of Travel and drop the former column


```

```python
travel_date=[]
for i in df['Date of Travel']:
    xl_date = i
    datetime_date = xlrd.xldate_as_datetime(xl_date, 0)
    date_object = datetime_date.date()
    travel_date.append(date_object)
```

```python
df['Travel_Date']= travel_date
df

#Drop Date of Travel Column:

df.drop ('Date of Travel', axis = 1 , inplace = True)
df
```

```python
#2. Calculate the company's income for each transaction and assign it to the dataframe
```

```python
df['income'] = df['Price Charged'] -df['Cost of Trip']
df
```

```python
#3. The cities ORANGE COUNTY and SILICON VALLEY need a STATE string for uniformity
#State : CA
```

```python
df['City'] = df['City'].str.replace('ORANGE COUNTY', 'ORANGE COUNTY CA')
df['City'] = df['City'].str.replace('SILICON VALLEY', 'SILICON VALLEY CA')
```

```python
df.loc[df['City'].isin (['SILICON VALLEY CA','ORANGE COUNTY CA'])]
```

```python
#4 . We create a function that extracts state initials from the city column and assign them to a new column called State
# another function that returns city without the state initials
```

```python
def stateGenerator (city):
    str = city.split(' ')
    a = len(str)
    if (a == 2):
        return str[1]
    else:
        return str[2]
    
        
```

```python
def stripCity (city):
    str = city.split(' ')
    a = len(str)
    if (a == 2):
        return str[0]
    else:
        return str[0] + ' ' + str[1]
```

```python
def segmentUsers(Income):
    if (Income > 0 and Income <= 9999):
        return 'LowerClass'
    elif (Income  > 9999 and Income <= 24999):
        return 'MiddleClass'
    else:
        return 'HighClass'
```

```python
df['Segment'] = df['Income (USD/Month)'].apply(segmentUsers)
```

```python
df['State'] = df['City'].apply(stateGenerator)
df['City'] = df['City'].apply(stripCity)

df
```

```python
#5. Extract the year and month columns for seasonal analysis of the data
```

```python
df['year'] = pd.DatetimeIndex(df['Travel_Date']).year
df['month'] = pd.DatetimeIndex(df['Travel_Date']).month
df
```
